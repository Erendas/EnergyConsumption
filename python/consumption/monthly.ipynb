```python
import base64

# monthly line plots for power usage dataset
from pandas import read_csv
from matplotlib import pyplot
import pandas as pd
# load the new file
dataset = read_csv( 'household_power_consumption.csv' , header=0, infer_datetime_format=True, parse_dates=['datetime'], index_col=['datetime'])
# plot active power for each month in the year 2006
year = 2006
pyplot.figure()

for month in range(1, 13):
    # prepare subplot
    ax = pyplot.subplot(12, 1, month)
    
    # determine the month to plot
    month_year = pd.Timestamp(year=year, month=month, day=1)  # Create datetime object
    
    # get all observations for the month in the year 2006
    result = dataset[(dataset.index.year == year) & (dataset.index.month == month)]
    
    # plot the active power for the month
    pyplot.plot(result.index, result['Global_active_power'])
    
    # add a title to the subplot
    pyplot.title(month_year.strftime('%Y-%m'), y=0, loc='left')  # Format month as 'YYYY-MM'
    
    # turn off ticks to remove clutter
    pyplot.yticks([])
    pyplot.xticks([])

pyplot.show()
```

```python
# monthly histogram plots for power usage dataset
from pandas import read_csv
from matplotlib import pyplot
# load the new file
dataset = read_csv( 'household_power_consumption.csv' , header=0, infer_datetime_format=True, parse_dates=['datetime'], index_col=['datetime'])
# plot active power for each year
year = 2006
pyplot.figure()
for month in range(1, 13):
  # prepare subplot
  ax = pyplot.subplot(12, 1, month)
  # determine the month to plot
  month_year = pd.Timestamp(year=year, month=month, day=1)
  # get all observations for the month
  result = dataset[(dataset.index.year == year) & (dataset.index.month == month)]
  # plot the active power for the month
  result['Global_active_power'].hist(bins=100)
  # zoom in on the distribution
  ax.set_xlim(0, 5)
  # add a title to the subplot
  pyplot.title(month_year.strftime('%Y-%m'), y=0, loc='right')  # Format month as 'YYYY-MM'
  # turn off ticks to remove clutter
  pyplot.yticks([])
  pyplot.xticks([])
pyplot.show()
```
